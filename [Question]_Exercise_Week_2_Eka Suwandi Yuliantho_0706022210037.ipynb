7. I have a character with this specification:
defense_overall 6.0<br>
offense_overall 7.0<br>
skill_effect_overall 4.0<br>
difficulty_overall 3.0<br>
movement_spd 250<br>
magic_defense 9<br>
mana 380<br>
hp_regen 37<br>
physical_atk 119<br>
physical_defense 24<br>
hp 2689<br>
attack_speed 0.75<br>
mana_regen 16.0<br>

Please predict its role using DecisionTree classification algorithm <br>

```python
# 7

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import metrics

# Encode the 'role' column
y = df6["role"]

# Separate features and target variable
x = df6[["defense_overall", "offense_overall", "skill_effect_overall", "difficulty_overall", "movement_spd",
        "magic_defense","mana","hp_regen","physical_atk","physical_defense","hp","attack_speed","mana_regen"]]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# Create a Decision Tree Classifier
decis = DecisionTreeClassifier()

# Train the model
decis = decis.fit(X_train, y_train)

# Make predictions on the test set
y_pred = decis.predict(X_test)

# Evaluate the model
print('Accuracy: ', metrics.accuracy_score(y_test, y_pred))

# Predict the role for the given character
predicted = decis.predict([[6.0, 7.0, 4.0, 3.0, 250, 9, 380, 37, 119, 24, 2689, 0.75, 16.0]])

# Get the predicted role (dicoding)
pred_role = ""
if predicted == 0:
    pred_role = "Fighter"
elif predicted == 1:
    pred_role = "tank"
elif predicted == 2:
    pred_role = "Support"
elif predicted == 3:
    pred_role = "Marksman"
elif predicted == 4:
    pred_role = "Mage"
elif predicted == 5:
    pred_role = "Assassin"
else:
    pred_role = "Unknown"


print("The Predicted Role for this guy is a: "+str(pred_role))
```
